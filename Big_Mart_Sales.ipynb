**Preprocessing Task before Model Building**

**Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df_train=df_train.apply(le.fit_transform)
df_train
```

```python
df_train.head()
```

**Spilting our data into train and test**

```python
!pip install sklearn
import sklearn
```

```python
import sklearn.model_selection as model_selection
```

```python
X=df_train.drop('item_outlet_sales',axis=1)
```

```python
y=df_train['item_outlet_sales']
```

```python
# split data into two sets 

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, train_size =0.75, test_size =0.25, random_state =101)
```

**Standardization**

```python
X.describe()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_test_std=sc.fit_transform(X_train)
```

```python
X_test_std
```

**Model Building**

```python
from sklearn.linear_model import LinearRegression
Ir=LinearRegression()
```

```python
Ir.fit(X_train,y_train)
```

```python
y_pred_Ir=Ir.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
print(r2_score(y_test,y_pred_Ir))
```

```python
print(mean_absolute_error(y_test,y_pred_Ir))
print(np.sqrt(mean_squared_error(y_test,y_pred_Ir)))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred_ref=rf.predict(X_test)
```

```python
print(r2_score(y_test,y_pred_ref))
```

```python
print(mean_absolute_error(y_test,y_pred_ref))
print(np.sqrt(mean_squared_error(y_test,y_pred_ref)))
```
